```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cleaned/zomato_cleaned.csv')
```

```python
# Set default theme
sns.set(style="whitegrid")
```

```python
top_locations = df['location'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(
    y=top_locations.index,
    x=top_locations.values,
    hue=top_locations.index,
    dodge=False,
    palette='viridis',
    legend=False
)
plt.title("Top 10 Locations with Most Restaurants in Bengaluru")
plt.xlabel("Number of Restaurants")
plt.ylabel("Location")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['rate'], bins=20, kde=True, color='orange')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['avg_cost_for_two'], bins=30, kde=True, color='green')
plt.title("Distribution of Average Cost for Two")
plt.xlabel("Cost (INR)")
plt.ylabel("Frequency")
plt.show()
```

```python
top_cuisines = df['cuisines'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(
    x=top_cuisines.values,
    y=top_cuisines.index,
    hue=top_cuisines.index,
    dodge=False,
    palette='rocket',
    legend=False
)
plt.title("Top 10 Most Popular Cuisines in Bengaluru")
plt.xlabel("Count")
plt.ylabel("Cuisines")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x='online_order', hue='online_order', palette='Set2', legend=False)
plt.title("Online Order Availability")

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='book_table', hue='book_table', palette='Set3', legend=False)
plt.title("Table Booking Availability")

plt.tight_layout()
plt.show()
```

```python
top_cities = df['city'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='coolwarm')
plt.title("Top Cities (Zomato Listing Frequency)")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```

```python
service_types = df['service_type'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=service_types.values, y=service_types.index, palette='magma')
plt.title("Distribution of Restaurant Service Types")
plt.xlabel("Count")
plt.ylabel("Service Type")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='service_type', y='rate', palette='pastel')
plt.title("Ratings by Service Type")
plt.xlabel("Service Type")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='online_order', y='avg_cost_for_two', palette='Set2')
plt.title("Cost for Two vs. Online Order Availability")
plt.xlabel("Online Order Available")
plt.ylabel("Avg Cost (INR)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='rate', y='votes', alpha=0.5)
plt.title("Votes vs. Ratings")
plt.xlabel("Rating")
plt.ylabel("Votes")
plt.show()
```

```python
top_rest_types = df['rest_type'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_rest_types.values, y=top_rest_types.index, palette='flare')
plt.title("Top 10 Restaurant Types")
plt.xlabel("Count")
plt.ylabel("Restaurant Type")
plt.show()
```
